## 중개중심성(betweeness) = 매개중심성
- 노드와 노드를 연결하는 역할을 하는 노드를 말한다. 
- 중개 중심성이 크다는 것은 그만큼 많은 노드의 사이에 위치하고 있다고 볼수 있고,
- 2개 노드를 연결하는 커뮤니케이터의 역할, 그리고 허브의 역할을 한다.

- 네트워크 내에서 한 노드가 담당하는 매개자 역할의 정도로서 중심성을 측정하는 방법입니다.

- A,B,C 세개의 노드가 있을때, A-B-C로 연결되어 있다고 가정하면, B는 A와 C의 매개자 역할을 합니다. 이때 C는 매개 중심성이 높다고 합니다.

- 즉, 한 노드가 연결망 내의 다른 노드들 사이의 최다 경로 위에 위치하면 할수록 그 노드의 매개 중심성이 높습니다.

```python
nx.betweenness_centrality(G2)
```

유해진 > 황정민 > 오달수 > 하정우 > 류준열

<br>

## 근접중심성(closeness)
- 근접중심성이 높은 값은 해당 노드에서 메세지가 발생했을 경우
- 노드 전체에 전달되는 최단거리가 가장 짧다

```python
nx.closeness_centrality(G2)
```

황정민 > 오달수 > 송강호

## 연결정도 중심성 (Degree centrality)
한 노드에 직접적으로 연결된 링크들의 합으로 얻어진 중심성 . 
일반적으로 쓰이는 Degree를 뜻합니다.

한 점에 얼마나 많은 다른 링크들이 관계를 맺고 있는지를 기준으로, 
노드의 중심 정도를 계량화 한 것입니다.

```python
nx.degree_centrality(G2)
```
