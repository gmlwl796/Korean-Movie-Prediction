# 1. 개봉전 7일 트윗

```python
before_tweets
```

```python
# 영화갯수가 262개가 맞는지 확인
len(before_tweets)/7
```

```python
# '명량'만 해보기
test = before_tweets['Frequency'][0:7].tolist()
test
sum(test)
```

```python
Before_Freq_sum = []
movie_name = []

for Freq in range(0, 262):
    Freq_7days = before_tweets['Frequency'][Freq*7:(Freq+1)*7].tolist()
    Freq_sum = sum(Freq_7days)
    Movie_name = before_tweets['Movie'][Freq*7]
    
    print('------------------------------')
    print(Freq_7days)    
    print(Freq_sum)
    print(Movie_name)
    
    Before_Freq_sum.append(Freq_sum)
    movie_name.append(Movie_name)
```

```python
print('총 영화 갯수 :', len(Before_Freq_sum))
print(Before_Freq_sum)
print('-------------------------------')
print('총 영화 갯수 :', len(movie_name))
print(movie_name)
```
