# 배급사와 영화배급수 데이터 

```python
import pandas as pd
bae = pd.read_csv('./csv2/bae.csv')
bae
```

```python
type(bae)
```

```python
'''
def bae(cnt):
    if cnt >= 30:
        print('대')
    elif cnt >= 10:
        print('중')
    else:
        print('소')
'''
```

```python
# 영화 갯수 컬럼 따로 변수로
cnt2 = bae.cnt
cnt2.head()
```

```python
cnt2[0]
```

```python
cnt_f = []

for i in range(len(cnt2)):
    if cnt2[i] >= 30:
        print('big')
        cnt_f.append('big')
    elif cnt2[i] >= 10:
        print('mid')
        cnt_f.append('mid')

    else:
        print('small')
        cnt_f.append('small')
```

```python
cnt_f
```

```python
cnt_f_df = pd.DataFrame(cnt_f, columns = ['크기'])
cnt_f_df
```

# cbind로 데이터 합치기

```python
bae_cnt_f = pd.concat([bae, cnt_f_df], axis=1)
bae_cnt_f
```

```python
bae_cnt_f.columns
```

# big, mid, small 로 리스트 나누기

```python
big = list(bae_cnt_f['배급사'][0:4])
big
```

```python
mid = list(bae_cnt_f['배급사'][4:8])
mid
```

```python
small = list(bae_cnt_f['배급사'][8:])
small
```

# 영화 데이터 가져오기

```python
import pandas as pd
movie = pd.read_csv('./csv2/movie8.csv')
movie = movie.drop('Unnamed: 0', axis=1)
movie['개봉일'] = movie['개봉일'].astype(str) 
movie.head()
```

```python
len(movie['배급사'])
```

```python
big
```

```python
movie['배급사크기'] = ''
movie
```

```python
for i in range(len(movie)):
    if movie['배급사'][i] in big:
        movie['배급사크기'][i] = 'big'
    elif movie['배급사'][i] in mid:
        if movie['배급사크기'][i] != 'big':
            movie['배급사크기'][i] = 'mid'
    elif movie['배급사'][i] in small:
        if movie['배급사크기'][i] == '':
            movie['배급사크기'][i] = 'small'

movie
```

```python
temp = movie[movie.배급사크기 == ''].index.values
for i in temp:
    tstr = movie['배급사'][i].split(',')
    for k in tstr:
        if movie['배급사크기'][i] != '':
            break
        elif k in big:
            movie['배급사크기'][i] = 'big'
        elif k in mid:
            movie['배급사크기'][i] = 'mid'
        elif k in small:
            movie['배급사크기'][i] = 'small'
        
```
