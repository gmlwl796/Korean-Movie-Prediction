```python
import pandas as pd
movie = pd.read_csv('./csv2/movie9.csv')
movie
```

```python
name_list = movie['영화명'].tolist()
name_list
```

```python
case1_name = movie.loc[(movie['영화명'].str.contains(':')),:]
case2_name = movie.loc[(movie['영화명'].str.contains('-')),:]
```

```python
target1 = case1_name['영화명'].tolist()
target2 = case2_name['영화명'].tolist()
print(target1)
print(target2)
```

```python
name_for_tweet = []
for name in target1:
    case1 = name.split(':')
    result = str.strip(case1[0])
    name_for_tweet.append(result)
name_for_tweet
```

```python
name_for_tweet2 = []
for name2 in target2:
    case2 = name2.split('-')
    result2 = str.strip(case2[0])
    name_for_tweet2.append(result2)
name_for_tweet2
```

```python
case1_name['영화명'] = name_for_tweet
case2_name['영화명'] = name_for_tweet2
```

```python
replace_target = case1_name.append(case2_name)
replace_target
```
