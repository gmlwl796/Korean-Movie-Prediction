### 두번째 모듈 사용 WordCloud

```python
import matplotlib.pyplot as plt
import nltk
from konlpy.tag import Twitter
t = Twitter()
%matplotlib inline

from wordcloud import WordCloud
```

```python
tokens_ko = t.nouns(strtemp)
tokens_ko
```

```python
tokens_ko2 = []
for i in range(len(tokens_ko)):
    if len(tokens_ko[i]) > 1:
        tokens_ko2.append(tokens_ko[i])
        
tokens_ko2
```

```python
ko = nltk.Text(tokens_ko2, name='극한직업2')
data = ko.vocab().most_common(500)
```

```python
temp_data = dict(data)
temp_data
```

```python
temp_data.pop('영화')
temp_data.pop('극한')
temp_data.pop('직업')
temp_data.pop('진짜')
temp_data.pop('보고')
temp_data.pop('오늘')
temp_data.pop('지금')
temp_data.pop('사람')
temp_data.pop('어제')
temp_data.pop('정말')
temp_data.pop('계속')
temp_data.pop('존나')
temp_data.pop('그냥')
temp_data.pop('여러분')
temp_data.pop('거의')
temp_data.pop('시작')
temp_data.pop('약간')
temp_data.pop('간다')
temp_data.pop('이름')
temp_data.pop('한국')
temp_data.pop('세상')
temp_data.pop('연예')
temp_data.pop('보기')
temp_data.pop('얼마나')
temp_data.pop('완전')
temp_data.pop('하루')
temp_data.pop('기분')
temp_data.pop('출처')
temp_data.pop('우리')
temp_data.pop('생각')
temp_data.pop('예매')
temp_data.pop('개봉')
temp_data.pop('부대')
temp_data.pop('무대')
temp_data.pop('인사')
temp_data['무대인사'] = 13
```

```python
temp_data
```

```python
wordcloud = WordCloud(font_path = 'C:\Windows\Fonts\Typo_SsangmunDongB.ttf',
                     relative_scaling = 0.2,
                     background_color='white',
                     ).generate_from_frequencies(temp_data)

plt.figure(figsize=(40,40))
plt.imshow(wordcloud)
plt.axis("off")
plt.savefig('./image/wordcloud_극한직업_트위터.jpg')
plt.show()
```

```python
import numpy as np
from PIL import Image
from wordcloud import ImageColorGenerator
korea_coloring = np.array(Image.open('./image/movie.jpg'))
korea_coloring
```

```python
image_colors = ImageColorGenerator(korea_coloring)
```

```python
plt.figure(figsize=(20,20))
plt.imshow(korea_coloring, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
wordcloud = WordCloud(font_path = 'C:\Windows\Fonts\Typo_SsangmunDongB.ttf',
                     relative_scaling = 0.2,
                     background_color='white',
                     mask = korea_coloring,
                     min_font_size = 1, max_font_size=60,
                     ).generate_from_frequencies(temp_data)

plt.figure(figsize=(10,10))
plt.imshow(wordcloud.recolor(color_func=image_colors), interpolation="bilinear")
#plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.savefig('./image/wordcloud_극한직업_with_image.jpg')
plt.show()
```
