## 아래 코드를 RUN 하기 전에 주의사항
#### Chromedriver.exe를 다운로드 받고 그 경로를 driver = 에 대입한다.(각자컴터)
#### 코드를 돌리면 크롬 팝업창이 뜨는데 건드리지 않는다.
#### 팝업창이 일시정지된것처럼 보이지만 건드리지 않는다.
#### (웹페이지 로딩시간을 일부러 준것임)
#### time sleep과 time wait를 임의로 줄이지 않는다. 
#### 크롤링이 안되는 경우 time sleep을 더 늘려준다. (인터넷환경이 느리면 타임슬립 더 길게 해야함)

```python
Director_names = mini_total_4['감독']
Movie_names = mini_total_4['영화명']
base_df = pd.DataFrame({"영화명":[0], "날짜":[0], "스크린수":[0], "상영횟수":[0], "매출액":[0], "관객수":[0],
                                   "누적매출액":[0], "누적관객수":[0], "순위":[0]}) 
for name in Movie_names:  
# 크롬 접속
    driver = webdriver.Chrome('C:/Users/student/Downloads/chromedriver_win32/chromedriver.exe')
    # URL 접속
    driver.get('http://www.kobis.or.kr/kobis/business/mast/mvie/searchMovieList.do')
    # ------조회할 정보 입력----------
    driver.implicitly_wait(15)
    # 영화 제목 입력
    driver.find_element_by_xpath('//*[@id="content"]/form[1]/fieldset/section/div[1]/table/tbody/tr[1]/td[1]/input').send_keys(str(name))
    # 영화 감독 입력
    driver.find_element_by_name('sDirector').send_keys('')
    # 날짜설정
    driver.find_element_by_name('sOpenYearS').send_keys('2014-01-01')
    driver.find_element_by_name('sOpenYearE').send_keys('2018-12-31')
    # 조회버튼
    next = driver.find_element_by_xpath('//*[@id="content"]/form[1]/fieldset/section/p/button[1]')
    next.click()
    # 영화 선택 및 통계자료 클릭
    driver.implicitly_wait(20)
    movie_1 = driver.find_element_by_xpath('//*[@id="content"]/table/tbody/tr[1]/td[1]/a')
    movie_1.click()
    driver.implicitly_wait(20)
    information = driver.find_element_by_xpath('/html/body/div[5]/div[1]/section/div/ul/li[2]/a')
    information.click()
    # 웹테이블 추출 (개봉후 n일의 스크린수가 있는 웹테이블)
    time.sleep(50)  # 시간이 걸리더라도 time.sleep 변경하지 말것!!!!
    # sleep 시간을 충분히 안주면 크롤링이 안됩니다!!
    tableScreen = driver.find_element_by_css_selector(".board02.bma.b02s.topico")
    driver.implicitly_wait(5)
    tableScreen.text
    tableScreen_content = []
    tableScreen_content = tableScreen.text.split()
    driver.implicitly_wait(5)
    if len(tableScreen_content) == 10 :
        tableScreen_df = pd.DataFrame({"영화명":[str(name)], "스크린수":[0], "상영횟수":[0], "매출액":[0], "관객수":[0], "누적매출액":[0], "누적관객수":[0], "순위":[0]})
    elif tableScreen_content[9] == 'N/A' :
            tableScreen_content = tableScreen_content[0:8] + tableScreen_content[10:]
            tableScreen_df = pd.DataFrame(np.array(tableScreen_content).reshape(int(len(tableScreen_content)/8), 8))
            tableScreen_df = tableScreen_df.rename(columns=tableScreen_df.iloc[0]).drop(tableScreen_df.index[0])
            new_order = [-1, 0, 1, 2, 3, 4, 5, 6, 7]
            tableScreen_df = tableScreen_df[tableScreen_df.columns[new_order]]
    else:
        tableScreen_df = pd.DataFrame(np.array(tableScreen_content).reshape(int(len(tableScreen_content)/8), 8))
        tableScreen_df = tableScreen_df.rename(columns=tableScreen_df.iloc[0]).drop(tableScreen_df.index[0])
        tableScreen_df['영화명'] = str(name)
        new_order = [-1, 0, 1, 2, 3, 4, 5, 6, 7]
        tableScreen_df = tableScreen_df[tableScreen_df.columns[new_order]]
    driver.close()
    globals()['movie_df_{}'.format(name)] = tableScreen_df
    print(globals()['movie_df_{}'.format(name)])
    base_df = base_df.append(tableScreen_df)
base_df
```

```python
final_df_3 = base_df.drop(0,0)
final_df_3
final_df_3.to_csv("final_df_3.csv")
```
