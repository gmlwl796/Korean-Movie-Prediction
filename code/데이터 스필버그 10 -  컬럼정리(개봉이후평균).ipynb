```python
movie['개봉평균스크린수'] = 0
movie['개봉평균상영횟수'] = 0
movie['개봉평균관객수'] = 0
```

```python
for i in range(len(movie)):
    movie['개봉평균스크린수'][i] = (movie['개봉1일_스크린수'][i] + movie['개봉2일_스크린수'][i] + movie['개봉3일_스크린수'][i]) / 3
    movie['개봉평균상영횟수'][i] = (movie['개봉1일_상영횟수'][i] + movie['개봉2일_상영횟수'][i] + movie['개봉3일_상영횟수'][i]) / 3
    movie['개봉평균관객수'][i] = (movie['개봉1일_관객수'][i] + movie['개봉2일_관객수'][i] + movie['개봉3일_관객수'][i]) / 3
```
