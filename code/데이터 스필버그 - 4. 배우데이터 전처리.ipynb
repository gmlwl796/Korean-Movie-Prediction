```python
import pandas as pd
import datetime
import requests

key1 = '93745210481f71d608b0ecd18fc8a7e9'

nullist = []
def crawl_data_actors(keys=key1):
    for i in range(0,len(actors)):
        url = 'http://www.kobis.or.kr/kobisopenapi/webservice/rest/people/searchPeopleList.json'

        payload = {
                    'key': keys, 
                   'peopleNm': actors['name'][i], 
                   'itemPerPage' : 50, 
                   'filmoNames' : actors['moviename'][i]
                  }

        req = requests.get(url, params=payload)
       
        if req.json()['peopleListResult']['totCnt'] == 1: # 1명이면 집어넣기
            actors['peopleCd'][i] = req.json()['peopleListResult']['peopleList'][0]['peopleCd']
            
        elif req.json()['peopleListResult']['totCnt'] > 1: # 여러명이면 돌려서 같은 사람 집어넣기
            for j in range(0, req.json()['peopleListResult']['totCnt']):
                if actors['moviename'][i] in req.json()['peopleListResult']['peopleList'][j]['filmoNames']:
                    actors['peopleCd'][i] = req.json()['peopleListResult']['peopleList'][j]['peopleCd']
                    
        else: # total == 0인 경우
            nullist.append(i) # 인덱스만 넣어줌

crawl_data_actors(key1)
print("완료")
```

# 수상내역 갯수 크롤링

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import urllib
import os
import sys
```

```python
peopleCd = actors['peopleCd']
prize_number = []
for Cd in peopleCd:
    url_base = "http://www.kobiz.or.kr/new/kor/01_bizmatching/movieInfo/searchPeopleView.jsp?peopleCd="
    url = url_base + str(Cd)
    req = requests.get(url)
    soup = BeautifulSoup(req.content.decode('utf-8','ignore'))
    soup.text
    if not '수상내역' in soup.text:  # 수상내역 테이블이 없으면 0 출력
        prize_number.append('0')
        print(url, '0')
    else: # 수상내역이 있으면 테이블 내에서 상 받은 것만 카운트
        contents = soup.select('div.boardview02.staffview > div.articleArea > div.h5Section > div > div > table > tbody > tr')
        contents
        dfcontent = []
        alldfcontents = []
        for content in contents:
            tds=content.find_all("td")
            for td in tds:
                dfcontent.append(td.text)
            alldfcontents.append(dfcontent)
            dfcontent = []
        content
        dfcontent
        alldfcontents
        df = pd.DataFrame(data = alldfcontents)
        count = df.iloc[:,3]!=''
        print(url, count.sum())
        prize_number.append(count.sum())
```

# 배우의 이전 상영작의 흥행여부 카운트 (기준: 5백만이상)

```python
import pandas as pd
actors = pd.read_csv('./csv2/actors_prize.csv', encoding='CP949')
actors
```

```python
temp2_cd = []
temp2_people = []
temp2_role = []
temp2_movie = []
total2 = []
```

```python
from bs4 import BeautifulSoup
import urllib.request

def countmovie(actors):
    cnt = 0
    for i in range(0, len(actors)):
        cnt = cnt + 1
        print(cnt)
        url = 'http://www.kobiz.or.kr/new/kor/01_bizmatching/movieInfo/searchPeopleView1.jsp?peopleCd=' + str(actors.peopleCd[i]) + '&pageIndex=1' 
        html = urllib.request.urlopen(url)
        soup = BeautifulSoup(html, 'html.parser')

        #movie_role = soup.findAll('dd', attrs={'class':'make'})
        movie_role = soup.select('dd.make')
        movie_name = soup.select('dt > a')
        movie_cnt2 = soup.select('dd > em.tl')
        temp = soup.select('a.last')
        print(temp)
        if len(temp) == 0:
            continue
        else:
            
            movie_page = temp[0].attrs['href']
            movie_page = movie_page[-2:-1]
            print(actors['name'][i] + movie_page + "/")
            
            for y in range(1, int(movie_page)+1):
                url = 'http://www.kobiz.or.kr/new/kor/01_bizmatching/movieInfo/searchPeopleView1.jsp?peopleCd=' + str(actors.peopleCd[i]) + '&pageIndex=' + str(y)
                html = urllib.request.urlopen(url)
                soup = BeautifulSoup(html, 'html.parser')
            
                movie_role = soup.findAll('dd', attrs={'class':'make'})
                movie_name = soup.select('dt > a')
                movie_cnt2 = soup.select('dd > em.tl')
                print(y)
                print(len(movie_name), len(movie_cnt2), len(movie_role))
            
                for j, t, k in zip(range(0, len(movie_name)),range(0, len(movie_role)), range(0, len(movie_cnt2), 2)): 
                    print(j, len(movie_name), k, len(movie_cnt2))
                
                    temp2_cd.append(actors.peopleCd[i])
                    temp2_people.append(actors.name[i])
                    temp2_role.append(movie_role[t].string)
                    temp2_movie.append(movie_name[j].string)
                    total2.append(str(movie_cnt2[k].string))                                                                              
                
countmovie(actors)
print('완료')
```
