```python
url= 'https://movie.naver.com/movie/bi/mi/pointWriteFormList.nhn?code=157297&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false&page={}'

import requests
import bs4
```

```python
res = requests.get(url.format(1))
```

```python
obj = bs4.BeautifulSoup(res.text)
```

```python
list1 = obj.find_all('div', {'class':'score_result'})
list1
```

```python
list2 = obj.find_all('div', {'class':'score_result'})[0].find_all('li')
list2
```

```python
list2[0].p.text
```

```python
score_result = dict()
count = 0
for li in list2:
    score_result[count] = {'score': int(li.em.text),
                          'reple' : li.p.text}
    count += 1
```

```python
score_result
```

```python
import pandas as pd
df = pd.DataFrame(score_result).T
```

```python
df
```

```python
df.to_csv('score_result.csv')
```

```python
count = 0
score_result = dict()

for page in range(0, 3):
    print(page)
    res = requests.get(url.format(page))
    obj = bs4.BeautifulSoup(res.text)
    print(obj)
    list2 = obj.find_all('div', {'class':'score_result'})[0].find_all('li')
    
    
    for li in list2:
        score_result[count] = {'score': int(li.em.text),
                          'reple': li.p.text}
        count += 1
    
df = pd.DataFrame(score_result).T
df.to_csv('score_result01.csv')
```
