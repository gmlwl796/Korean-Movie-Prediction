# 영화인 API에서 감독 정보 추출 - 감독테이블 생성

```python
import pandas as pd

movie = pd.read_csv('./csv2/movie.csv')
movie
```

```python
directors = []
directors = pd.DataFrame(directors)
directors['directors'] = movie['감독'].copy()
directors['movieNm'] = movie['영화명'].copy()
directors['peopleCd'] = '0'
```

```python
import pandas as pd
import datetime
import requests

key1 = '93745210481f71d608b0ecd18fc8a7e9'

nullist = []
def crawl_data_directors(keys=key1):
    for i in range(0,len(directors)):
        url = 'http://www.kobis.or.kr/kobisopenapi/webservice/rest/people/searchPeopleList.json'

        payload = {
                    'key': keys, 
                   'peopleNm': movie['감독'][i], 
                   'itemPerPage' : 50, 
                   'filmoNames' :  movie['영화명'][i]
                  }

        req = requests.get(url, params=payload)
        if req.json()['peopleListResult']['totCnt'] == 1: # 1명이면 집어넣기
            directors['peopleCd'][i] = req.json()['peopleListResult']['peopleList'][0]['peopleCd']
        elif req.json()['peopleListResult']['totCnt'] > 1: # 여러명이면 돌려서 같은 사람 집어넣기
            for j in range(0, req.json()['peopleListResult']['totCnt']):
                if movie['영화명'][i] in req.json()['peopleListResult']['peopleList'][j]['filmoNames']:
                    directors['peopleCd'][i] = req.json()['peopleListResult']['peopleList'][j]['peopleCd']
                    break
        else: # total == 0인 경우
            nullist.append(i) # 인덱스만 넣어줌


crawl_data_directors(keys=key1)

print("완료")
```

# 수상내역 갯수 크롤링

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import urllib
import os
import sys
```

```python
test1 = directors['peopleCd'][0:50]
test2 = directors['peopleCd'][50:100]
test3 = directors['peopleCd'][100:150]
test4 = directors['peopleCd'][150:200]
test5 = directors['peopleCd'][200:210]
test5_6 = directors['peopleCd'][210:211]
test6 = directors['peopleCd'][211:262]
```

```python
peopleCd = test6
prize_number = []
for Cd in peopleCd:
    url_base = "http://www.kobiz.or.kr/new/kor/01_bizmatching/movieInfo/searchPeopleView.jsp?peopleCd="
    url = url_base + str(Cd)
    req = requests.get(url)
    soup = BeautifulSoup(req.content.decode('utf-8','ignore'))
    soup.text
    if not '수상내역' in soup.text:  # 수상내역 테이블이 없으면 0 출력
        prize_number.append(0)
        print(url, 0)
    else: # 수상내역이 있으면 테이블 내에서 상 받은 것만 카운트
        contents = soup.select('div.boardview02.staffview > div.articleArea > div.h5Section > div > div > table > tbody > tr')
        contents
        dfcontent = []
        alldfcontents = []
        for content in contents:
            tds=content.find_all("td")
            for td in tds:
                dfcontent.append(td.text)
            alldfcontents.append(dfcontent)
            dfcontent = []
        content
        dfcontent
        alldfcontents
        df = pd.DataFrame(data = alldfcontents)
        count = df.iloc[:,3]!=''
        print(url, count.sum())
        prize_number.append(count.sum())
```

```python
directors['prizecount'][211:262] = prize_number
```

# 감독 필모그라피에서 누적관객수 

```python
import pandas as pd
directors = pd.read_csv("./csv2/directors_prize_final.csv")
directors
```

```python
temp2_cd = []
temp2_people = []
temp2_role = []
temp2_movie = []
total2 = []
directors.peopleCd = directors.peopleCd.astype(str) 
```

```python
from bs4 import BeautifulSoup
import urllib.request

        
def countmovie(directors):
    for i in range(0, len(directors)):
        url = 'http://www.kobiz.or.kr/new/kor/01_bizmatching/movieInfo/searchPeopleView1.jsp?peopleCd=' + directors.peopleCd[i] + '&pageIndex=1' 
        html = urllib.request.urlopen(url)
        soup = BeautifulSoup(html, 'html.parser')

        #movie_role = soup.findAll('dd', attrs={'class':'make'})
        movie_role = soup.select('dd.make')
        movie_name = soup.select('dt > a')
        movie_cnt2 = soup.select('dd > em.tl')
        temp = soup.select('a.last')
        print(temp)
        movie_page = temp[0].attrs['href']
        movie_page = movie_page[-2:-1]
        print(directors['directors'][i] + movie_page + "/")
        
        for y in range(1, int(movie_page)+1):
            url = 'http://www.kobiz.or.kr/new/kor/01_bizmatching/movieInfo/searchPeopleView1.jsp?peopleCd=' + directors.peopleCd[i] + '&pageIndex=' + str(y)
            html = urllib.request.urlopen(url)
            soup = BeautifulSoup(html, 'html.parser')
            
            movie_role = soup.findAll('dd', attrs={'class':'make'})
            movie_name = soup.select('dt > a')
            movie_cnt2 = soup.select('dd > em.tl')
            print(y)
            print(len(movie_name), len(movie_cnt2), len(movie_role))
            
            for j, t, k in zip(range(0, len(movie_name)),range(0, len(movie_role)), range(0, len(movie_cnt2), 2)): 
                print(j, len(movie_name), k, len(movie_cnt2))
                
                temp2_cd.append(directors.peopleCd[i])
                temp2_people.append(directors.directors[i])
                temp2_role.append(movie_role[t].string)
                temp2_movie.append(movie_name[j].string)
                total2.append(str(movie_cnt2[k].string))
                
countmovie(directors)
print('완료')
```

```python
print(len(temp2_cd), len(temp2_people), len(temp2_role), len(temp2_movie), len(total2))
```

```python
directors2 = pd.DataFrame(columns=['directors','peopleCd', 'role', 'movie', 'totalpeople'])
directors2['directors'] = temp2_people
directors2['peopleCd'] = temp2_cd
directors2['role'] = temp2_role
directors2['movie'] = temp2_movie
directors2['totalpeople'] = total2
directors2
```

```python
for i in range(0, len(directors2)):
    if '감독' != directors2['role'][i]:
        directors2 = directors2.drop([i])
directors2.head()
```

```python
directors2 = directors2.reset_index()
directors2.head()
```

```python
directors2 = directors2.drop(['index'], axis=1)
directors2.head()
```

```python
for i in range(len(directors2)):
    directors2['totalpeople'][i] = str(directors2['totalpeople'][i])
    directors2['totalpeople'][i] = directors2['totalpeople'][i].replace(',','')
    if '정' in directors2['totalpeople'][i]:
        directors2['totalpeople'][i] = -1
    directors2['totalpeople'][i] = int(directors2['totalpeople'][i])
```
