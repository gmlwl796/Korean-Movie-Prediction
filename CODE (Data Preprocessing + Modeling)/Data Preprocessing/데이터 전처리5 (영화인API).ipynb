```python
import pandas as pd

movie_list = pd.read_csv("./csv/final.csv", encoding= 'CP949')
movie_list.head()
```

```python
directors = []
directors = pd.DataFrame(directors)
directors['directors'] = movie_list['directors'].copy()
directors['movieNm'] = movie_list['movieNm'].copy()
directors['peopleCd'] = '0'
directors['Role'] = '-'
```

```python
import pandas as pd
import datetime
import requests

key1 = '93745210481f71d608b0ecd18fc8a7e9'

nullist = []
def crawl_data(keys=key1):
    for i in range(0,len(directors)):
        url = 'http://www.kobis.or.kr/kobisopenapi/webservice/rest/people/searchPeopleList.json'

        payload = { 'key': keys, 'peopleNm': movie_list['directors'][i], 'itemPerPage' : 50, 'filmoNames' :  movie_list['movieNm'][i]}

        req = requests.get(url, params=payload)
       #print(type(directors['movieNm'].astype(str)), type(req.json()['peopleListResult']['peopleList'][0]['filmoNames']))
        print(req.json()['peopleListResult'])
        
        if req.json()['peopleListResult']['totCnt'] == 1: # 1명이면 집어넣기
            directors['peopleCd'][i] = req.json()['peopleListResult']['peopleList'][0]['peopleCd']
            directors['Role'][i] = req.json()['peopleListResult']['peopleList'][0]['repRoleNm']
        elif req.json()['peopleListResult']['totCnt'] > 1: # 여러명이면 돌려서 같은 사람 집어넣기
            for j in range(0, req.json()['peopleListResult']['totCnt']):
                if movie_list['directors'][i] in req.json()['peopleListResult']['peopleList'][j]['filmoNames']:
                    directors['peopleCd'][i] = req.json()['peopleListResult']['peopleList'][j]['peopleCd']
                    directors['Role'][i] = req.json()['peopleListResult']['peopleList'][j]['repRoleNm']
        else: # total == 0인 경우
            nullist.append(i) # 인덱스만 넣어줌
            payload = { 'key': keys, 'peopleNm': movie_list['directors'][i], 'itemPerPage' : 50, 'filmoNames' :  movie_list['movieNm'][i][0:3]}
            req = requests.get(url, params=payload)
            if req.json()['peopleListResult']['totCnt'] == 1: # 1명이면 집어넣기
                directors['peopleCd'][i] = req.json()['peopleListResult']['peopleList'][0]['peopleCd']
                directors['Role'][i] = req.json()['peopleListResult']['peopleList'][0]['repRoleNm']
            elif req.json()['peopleListResult']['totCnt'] > 1: # 여러명이면 돌려서 같은 사람 집어넣기
                for j in range(0, req.json()['peopleListResult']['totCnt']):
                    if movie_list['directors'][i] in req.json()['peopleListResult']['peopleList'][j]['filmoNames']:
                        directors['peopleCd'][i] = req.json()['peopleListResult']['peopleList'][j]['peopleCd']
                        directors['Role'][i] = req.json()['peopleListResult']['peopleList'][j]['repRoleNm']


crawl_data(keys=key1)
```
