# 특정 기간에 상영된 영화에 대한 상세정보를 API를 이용하여 추출하기

```python
import pandas as pd

movie_list = pd.read_csv("data1.csv", encoding='CP949')
movie_list
```

```python
movie_list.movieCd = movie_list.movieCd.astype(str)
```

```python
import json
import requests

key2 = 'e7f4a98d57f2e4cba802338e862b2919'

def crawl_movie_data(lists, keys = key2) :
    
    dict_list = dict()
    url = 'http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieInfo.json'
    
    for movie in lists :
        print(movie)
        payload={
        'key' : keys,
        'movieCd' : movie
        }
        
        req = requests.get(url, params = payload)
        
        ## 각 영화 코드가 key, 그 결과값으로 나온 json 항목들이 value가 된다.
        dict_list[movie] = req.json()['movieInfoResult']['movieInfo']

    return dict_list
   
movie_detail = crawl_movie_data(movie_list.movieCd)

with open('detail.txt', 'w') as outfile:  
    json.dump(movie_detail, outfile)
```
