```python
a=[]


movieCd = totalmovie['movieCd']
for Cd in movieCd:
    base = 'https://movie.naver.com/movie/bi/mi/point.nhn?code='
    url = base + str(Cd)
    req = requests.get(url)
    html = req.text
    soup = BeautifulSoup(html, 'html.parser')
    movie_grade = soup.select('#content > div.article > div.mv_info_area > div.mv_info > dl > dd > p > a')
    for grade in movie_grade:
        movie_grade = grade.text
        a.append(movie_grade)
```
