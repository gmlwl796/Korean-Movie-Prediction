### 두번째 모듈 사용 WordCloud

```python
import matplotlib.pyplot as plt
import nltk
from konlpy.tag import Twitter
t = Twitter()
%matplotlib inline

from wordcloud import WordCloud
```

```python
text = open("./csv2/movie_review.txt").read()
tokens_ko = t.nouns(text)
tokens_ko
```

```python
tokens_ko2 = []
for i in range(len(tokens_ko)):
    if len(tokens_ko[i]) > 1:
        tokens_ko2.append(tokens_ko[i])
        
tokens_ko2
```

```python
ko = nltk.Text(tokens_ko2, name='마약왕')
data = ko.vocab().most_common(500)
```

```python
temp_data = dict(data)
temp_data
```

```python
temp_data.pop('영화')
temp_data.pop('관람객')
temp_data.pop('진짜')
temp_data.pop('그냥')
temp_data.pop('생각')
temp_data.pop('정도')
temp_data.pop('하나')
temp_data.pop('때문')
temp_data.pop('이영화')
temp_data.pop('내용')
temp_data.pop('느낌')
temp_data.pop('정말')
temp_data.pop('부분')
temp_data.pop('보고')
temp_data.pop('가지')
temp_data.pop('중간')
temp_data.pop('한국영')
temp_data.pop('모두')
temp_data.pop('사실')
temp_data.pop('이건')
temp_data.pop('우민')
temp_data.pop('갑자기')
temp_data.pop('전체')
temp_data.pop('한번')
temp_data.pop('이번')
temp_data.pop('표정')
temp_data.pop('살짝')
temp_data.pop('언제')
temp_data.pop('다시')
temp_data.pop('요소')
```

```python
temp_data
```

```python
wordcloud = WordCloud(font_path = 'C:\Windows\Fonts\Typo_SsangmunDongB.ttf',
                     relative_scaling = 0.2,
                     background_color='white',
                     ).generate_from_frequencies(temp_data)

plt.figure(figsize=(40,40))
plt.imshow(wordcloud)
plt.axis("off")
plt.savefig('./image/wordcloud_마약왕_네이버.jpg')
plt.show()
```

```python
import numpy as np
from PIL import Image
from wordcloud import ImageColorGenerator
korea_coloring = np.array(Image.open('./image/movie.jpg'))
korea_coloring
```

```python
image_colors = ImageColorGenerator(korea_coloring)
```

```python
plt.figure(figsize=(20,20))
plt.imshow(korea_coloring, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
wordcloud = WordCloud(font_path = 'C:\Windows\Fonts\Typo_SsangmunDongB.ttf',
                     relative_scaling = 0.2,
                     background_color='white',
                     mask = korea_coloring,
                     min_font_size = 1, max_font_size=60,
                     ).generate_from_frequencies(temp_data)

plt.figure(figsize=(10,10))
plt.imshow(wordcloud.recolor(color_func=image_colors), interpolation="bilinear")
#plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.savefig('./image/wordcloud_마약왕_with_image.jpg')
plt.show()
```
