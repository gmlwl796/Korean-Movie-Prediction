```r
#install.packages('XML')
#install.packages('stringr')
library(XML)
library(stringr)
```

```r
all_reviews <- NULL
url_base <- "http://movie.naver.com/movie/bi/mi/pointWriteFormList.nhn?code=166092&type=after&onlyActualPointYn=N&order=newest&page="  
# 17421 는 쇼생크 탈출이에여 
# 이거를 다른 영화 url의 고유숫자로 바꾸면됩니당.

# for (i in 1:100)에서
# 100은 100번째 리뷰 페이지입니다
# 예를들어 두번째 페이지까지만 하고 싶으면 2 적으면 돼요
for (i in 1:100) {
  temp <- NULL
  url <- paste(url_base, i, sep='')
  txt <- readLines(url, encoding="UTF-8")
  
  points <- txt[which(str_detect(txt, "class=\"star_score\""))+2]
  points <- gsub("<.+?>|\t", "", points)
  points <- points[c(2:11)]  # 한 리뷰 페이지당 데이터가 11건이 나옴 (첫번째의 불필요한행 존재. 때문에 2번째 행부터 11번째 행의 값만 가져옴)
  
  reviews <- txt[which(str_detect(txt, "class=\"score_reple\"")) + 1]
  reviews <- gsub("<.+?>|\t", "", reviews)
  
  temp <- cbind(points, reviews)  #각 페이지의 10건 리뷰를 temp 변수에 담습니다.
  all_reviews <- rbind(all_reviews, temp)
}

write.table(all_reviews, "./csv2/movie_review_pmc.txt")  # C드라이브 Temp 폴더에 저장
```
