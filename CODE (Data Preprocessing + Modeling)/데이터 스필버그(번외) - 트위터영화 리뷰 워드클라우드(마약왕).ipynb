### 두번째 모듈 사용 WordCloud

```python
import matplotlib.pyplot as plt
import nltk
from konlpy.tag import Twitter
t = Twitter()
%matplotlib inline

from wordcloud import WordCloud
```

```python
tokens_ko = t.nouns(strtemp)
tokens_ko
```

```python
tokens_ko2 = []
for i in range(len(tokens_ko)):
    if len(tokens_ko[i]) > 1:
        tokens_ko2.append(tokens_ko[i])
        
tokens_ko2
```

```python
ko = nltk.Text(tokens_ko2, name='마약왕2')
data = ko.vocab().most_common(200)
```

```python
temp_data = dict(data)
temp_data
```

```python
import numpy as np
from PIL import Image
from wordcloud import ImageColorGenerator
korea_coloring = np.array(Image.open('./image/movie.jpg'))
korea_coloring
```

```python
image_colors = ImageColorGenerator(korea_coloring)
```

```python
plt.figure(figsize=(20,20))
plt.imshow(korea_coloring, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
wordcloud = WordCloud(font_path = 'C:\Windows\Fonts\Typo_SsangmunDongB.ttf',
                     relative_scaling = 0.2,
                     background_color='white',
                     mask = korea_coloring,
                     min_font_size = 1, max_font_size=60,
                     ).generate_from_frequencies(temp_data)

plt.figure(figsize=(10,10))
plt.imshow(wordcloud.recolor(color_func=image_colors), interpolation="bilinear")
#plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.savefig('./image/wordcloud_마약왕트위터_with_image.jpg')
plt.show()
```
